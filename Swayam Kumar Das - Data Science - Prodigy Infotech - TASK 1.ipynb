```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("population.csv")
data.head(266)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'data' is already loaded
# List of countries for the graph
countries_to_plot = ['United States', 'Canada', 'Brazil', 'Japan', 'Germany', 'France']

# Filtering the data frame
filtered_data = data[data['Country Name'].isin(countries_to_plot)]

# Sort data by population for better visualization
filtered_data = filtered_data.sort_values('2022', ascending=False)

# Set the style and color palette
sns.set_style("whitegrid")
sns.set_palette("deep")

# Create the figure and axis objects
plt.figure(figsize=(12, 6))

# Create the bar chart
ax = sns.barplot(x='Country Name', y='2022', data=filtered_data)

# Customize the chart
plt.title('Population of Selected Countries in 2022', fontsize=16, fontweight='bold')
plt.xlabel('Country', fontsize=12)
plt.ylabel('Population', fontsize=12)
plt.xticks(rotation=45, ha='right')

# Add value labels on top of each bar
for i, v in enumerate(filtered_data['2022']):
    ax.text(i, v, f'{v:,.0f}', ha='center', va='bottom', fontweight='bold')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
